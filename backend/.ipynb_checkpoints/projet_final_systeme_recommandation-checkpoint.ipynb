```python
import pandas as pd
df1 = pd.read_csv("./student-scores.csv")
df = df1.copy()
df.head()
```

```python
df.columns
df.drop(columns=['id','first_name','last_name','email'],axis=1, inplace=True)
```

```python
df["total_score"] = df["math_score"] + df["history_score"] + df["physics_score"] + df["chemistry_score"] + df["biology_score"] + df["english_score"] + df["geography_score"]
df["average_score"] = df["total_score"] / 7
df.head()
```

```python
# Supprimer les lignes où 'career_aspiration' est égal à 'Unknown'
df = df[df['career_aspiration'] != 'Unknown']

# Vérifier le résultat
print(df['career_aspiration'].unique())
print(len(df))  # Affiche le nouveau nombre de lignes
```

```python
df
```

```python
career_to_field_map = {
    'Lawyer': 'Droit',
    'Doctor': 'Médecine',
    'Government Officer': 'Sciences politiques',
    'Artist': 'Beaux-arts',
    'Software Engineer': 'Informatique',
    'Teacher': 'Éducation',
    'Business Owner': 'Entrepreneuriat',
    'Scientist': 'Recherche scientifique',
    'Banker': 'Finance',
    'Writer': 'Littérature',
    'Accountant': 'Comptabilité',
    'Designer': 'Design',
    'Construction Engineer': 'Génie civil',
    'Game Developer': 'Développement de jeux',
    'Stock Investor': 'Économie et marchés financiers',
    'Real Estate Developer': 'Immobilier'
}
```

```python
df
```

```python
print(df.columns)
```

```python
# Remplacer la colonne 'career_aspiration' par la nouvelle colonne 'Field'
df['career_aspiration'] = df['career_aspiration'].map(career_to_field_map)

# Renommer la colonne 'career_aspiration' en 'Field' si vous voulez la remplacer complètement
df.rename(columns={'career_aspiration': 'Field'}, inplace=True)
```

```python
df
```

Encoding Categorical Columns

```python
gender_map = {'male': 0, 'female': 1}
part_time_job_map = {False: 0, True: 1}
extracurricular_activities_map = {False: 0, True: 1}
Field_map = {
        'Droit': 0, 'Médecine': 1, 'Sciences politiques': 2, 'Beaux-arts': 3, 'Informatique': 4,
      'Non spécifié': 5,'Éducation': 6, 'Entrepreneuriat': 7, 'Recherche scientifique': 8,
        'Finance': 9, 'Littérature': 10, 'Comptabilité': 11, 'Design': 12,
        'Génie civil': 13, 'Développement de jeux': 14, 'Économie et marchés financiers': 15,
        'Immobilier': 16
    }
# Apply mapping to the DataFrame
df['gender'] = df['gender'].map(gender_map)
df['part_time_job'] = df['part_time_job'].map(part_time_job_map)
df['extracurricular_activities'] = df['extracurricular_activities'].map(extracurricular_activities_map)
df['Field'] = df['Field'].map(Field_map)
```

```python
df['Field'].unique()
```

```python
df
```

```python
df.columns
```

```python
df.drop(columns=['gender','part_time_job','extracurricular_activities'],axis=1, inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

# Encoder la variable cible si elle est catégorique
y = LabelEncoder().fit_transform(df['Field'])

# Encoder les colonnes catégoriques de X
X = pd.get_dummies(df.drop('Field', axis=1))
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Encoder la variable cible
y = LabelEncoder().fit_transform(df['Field'])

# Encoder les colonnes catégoriques dans X
X = pd.get_dummies(df.drop('Field', axis=1))

# Vérifiez les déséquilibres dans y
print("Répartition des classes avant SMOTE :")
print(pd.Series(y).value_counts())

# Créez un objet SMOTE
smote = SMOTE(random_state=42)

# Appliquez SMOTE
X_resampled, y_resampled = smote.fit_resample(X, y)

# Vérifiez les nouvelles proportions
print("Répartition des classes après SMOTE :")
print(pd.Series(y_resampled).value_counts())
```

# Train test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_resampled,y_resampled,test_size=0.2, random_state=42)
```

```python
X_train.shape
```

# Feature Scalling

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler to the training data and transform both training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled.shape
```

# Models Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import warnings
warnings.filterwarnings("ignore")

# Define models
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest Classifier": RandomForestClassifier(),
    "K Nearest Neighbors": KNeighborsClassifier(),
  
}

# Train and evaluate each model
for name, model in models.items():
    print("="*50)
    print("Model:", name)
    # Train the model
    model.fit(X_train_scaled, y_train)
    
    # Predict on test set
    y_pred = model.predict(X_test_scaled)
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    classification_rep = classification_report(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    
    # Print metrics
    print("Accuracy:", accuracy)
    print("Classification Report:\n", classification_rep)
    print("Confusion Matrix:\n", conf_matrix)
```

```python
model = RandomForestClassifier()

model.fit(X_train_scaled, y_train)
# Predict on test set
y_pred = model.predict(X_test_scaled)

# Calculate metrics
print("Accuracy: ",accuracy_score(y_test, y_pred))
print("Report: ",classification_report(y_test, y_pred))
print("Confusion Matrix: ",confusion_matrix(y_test, y_pred))
```

# Single Input Predictions

```python
#test1
print("Actual Label:", y_test[10])  # Utilisez l'indexation de tableau
print("Model Prediction:", model.predict(X_test_scaled[10].reshape(1, -1))[0])

if y_test[10] == model.predict(X_test_scaled[10].reshape(1, -1)):
    print("Prediction is correct!")
else:
    print("Prediction is incorrect.")
```
